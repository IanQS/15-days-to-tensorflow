```python
import numpy as np
import tensorflow as tf
```

```python
from tensorflow.examples.tutorials.mnist import input_data
```

```python
data=input_data.read_data_sets("/tmp/data/",one_hot=True)
```

```python
x_train=data.train.images
y_train=data.train.labels
```

```python
x_test=data.test.images
y_test=data.test.labels
```

```python
cell =tf.nn.rnn_cell.LSTMCell(num_units=512,)
```

```python
#hyperparameters

time_step=28
input_dim=28
lr=0.001
n_class=10
batch_size=128
```

```python
weight_forward=tf.Variable(tf.random_normal([512,n_class]))
```

```python
bias_forward=tf.Variable(tf.random_normal([n_class]))
```

```python
#placeholders

x=tf.placeholder(float,[None,28,28])
y=tf.placeholder(float,[None,10])
```

```python
input=tf.unstack(x,28,axis=1)
```

```python
h_output,h_hidden=tf.contrib.rnn.static_rnn(cell,input,dtype="float32")
```

```python
prediction=tf.matmul(h_output[-1],weight_forward)+bias_forward
prediction.shape
```

```python
loss=tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=prediction,labels=y))
opt=tf.train.AdamOptimizer(learning_rate=lr).minimize(loss)

#model evaluation
correct_prediction=tf.equal(tf.argmax(prediction,1),tf.argmax(y,1))
accuracy=tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
```

```python
init=tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    iter=1
    while iter<3000:
        batch_x,batch_y=data.train.next_batch(batch_size=batch_size)
        
        batch_x=batch_x.reshape((batch_size,28,28))

        sess.run(opt, feed_dict={x: batch_x, y: batch_y})

        if iter %10==0:
            acc=sess.run(accuracy,feed_dict={x:batch_x,y:batch_y})
            los=sess.run(loss,feed_dict={x:batch_x,y:batch_y})
            print("For iter ",iter)
            print("Accuracy ",acc)
            print("Loss ",los)
            print("__________________")

        iter=iter+1
    test_data = data.test.images[:128].reshape((-1, 28,28))
    test_label = data.test.labels[:128]
    print("Testing Accuracy:", sess.run(accuracy, feed_dict={x: test_data, y: test_label}))
```
